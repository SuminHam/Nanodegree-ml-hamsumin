# 1.1. 디자인패턴 <br>

프로그램을 설계할 때 발생했던 문제점들을 객체 간의 상호 관계 등을 이용하여 해결할 수 있도록 하나의 '규약' 형태로 만들어 놓은 것.<br>

쉽게 말하자면 일반적으로 발생하는 문제에 대한 일반적으로 재사용 가능한 솔루션.<br>

크게 세가지 패턴으로 분류된다:
* 생성 패턴: 객체 생성 메커니즘에 중점을 두고 상황에 적합한 방식으로 객체를 생성할 수 있는 방법을 제공.
* 구조적 패턴: 객체 구성과 관련하여 클래스와 객체를 결합하여 더 큰 구조를 형성하는 방법을 정의.
* 행동 패턴: 개체 협업 및 책임에 중점을 두고 개체가 서로 상호 작용하고 통신하는 방식을 정의.

## 1.1.1. 싱글톤 패턴<br>

하나의 클래스에 하나의 인스턴스만 가지는 패턴.<br><br>


### 싱글톤 패턴의 특징<br>

* 하나의 인스턴스를 만들어 놓고 해당 인스턴스를 다른 모듈들이 공유하며 사용하기 때문에 인스터스를 생성할 때 드는 비용이 줄어든다.<br>
  
* 시스템 전체에 클래스의 인스턴스가 하나만 있기 때문에 싱글톤은 애플리케이션에 전역 상태를 도입하기 때문에 의존성이 높다.<br>

* 의존성 주입: 메인 모듈이 의존성 주입자를 도입해 간접적으로 하위 모듈에 대해 의존성을 주입할수 있으며, 이를 디커플링이 된다라고 한다.<br>

### 싱글톤 패턴의 단점<br>

TDD(Test Driven Development)를 할 때 걸림돌이 된다. 단위 테스트는 테스트가 서로 독립적이어야 하는데, 싱글톤 패턴은 각 테스트마다 독립적인 인스턴스를 만들기가 어렵기 때문이다.<br>

### 의존성 주입의 장단점<br>

* 장점: 모듈들을 쉽게 교체할 수 있는 구조가 되어 테스팅하기 쉽고 마이그레이션 하기도 수월하다.<br>
 또한 구현할 때 추상화 레이어를 넣고 이를 기반으로 구현체를 넣어 주기 때문에 애플리케이션 의존성 방향이 일관되고, 애플리케이션을 쉽게 추론할 수 있으며, 모듈 간의 관계들이 조금 더 명확해진다.<br>

* 단점: 모듈들이 더욱더 분리되므로 클래스 수가 늘어나 복잡성이 증가될 수 있으며 약간의 런타임 페널티가 생기기도 한다. <br><br><br>



## 1.1.2 팩토리 패턴<br>

객체를 사용하는 코드에서 객체 생성 부분을 떼어내 추상화한 패턴이자 상속 관계에 있는 두 클래스에서 상위 클래스가 중요한 뼈대를 결정하고, 하위클래스에서 객체 생성에 관한 구체적인 내용을 결정하는 패턴이다.<br>

상위 클래스에서는 인스턴스 생성 방식에 대해 전혀 알 필요가 없어 더 많은 유연성을 가진다.<br>
또한 객체 생성 로직이 따로 뗴어져 있기 때문에 코드를 리팩터링하더라도 한 곳만 고칠 수 있게 되닌 유지 보수성이 증가한다.<br>

다음은 팩토리 패턴에 대한 예시이다:
>단순화된 피자 가게에서 작동하는 방법은 다음과 같습니다.<br><br>
>고객: 피자가 어떻게 만들어졌는지, 누가 만드는지 알 필요 없이 피자를 주문합니다.<br><br>
>피자 셰프(크리에이터): 일반적인 피자 제작 과정을 따릅니다. 그들은 어떤 피자를 만들 것인지 결정하지 않습니다. 그들은 그 결정을 전문 요리사에게 맡깁니다.<br><br>
>전문 피자 셰프(콘크리트 크리에이터): 각 셰프는 특정 유형의 피자를 전문적으로 만듭니다. 표준 피자 제조 과정에서 피자가 필요할 때 그들은 전문 피자를 만듭니다.<br><br>
>본질적으로 팩토리 패턴을 사용하면 객체의 정확한 클래스를 지정하지 않고도 객체를 생성할 수 있습니다. 우리는 객체를 생성하는 표준 방법을 가지고 있지만, 다양한 하위 클래스는 생성 프로세스를 사용자 정의하여 다양한 유형의 객체(우리의 경우 피자)를 생성할 수 있습니다. 이러한 유연성 덕분에 기존 코드를 변경하지 않고도 새로운 유형의 개체(피자)를 더 쉽게 추가할 수 있습니다.<br><br>

## 1.1.3. 전략패턴<br>

정책 패턴이라고도 하며, 객체의 행위를 바꾸고 싶은 경우 '직접' 수정하지 앟고 전략이라고 부르는 '캡슐화한 알고리즘'을 컨텍스트 안에서 바꿔주면서 상호 교체가 가능하게 만드는 패턴이다.<br>

다음은 전략패턴에 대한 간단한 예시이다:<br>
> 컨텍스트(내 캐릭터): 적과 마주치면 그들을 물리쳐야 합니다. 고정된 전투 방식이 아닌 상황에 따라    다양한 무기를 전환할 수 있습니다.<br><br>
> 전략 인터페이스(무기): attack() 또는 defend()와 같이 모든 무기가 구현해야 하는 메서드를 정의합니다. 모든 무기(전략)에는 이러한 메소드가 있어야 합니다.<br><br>
> 구체적인 전략(검, 활, 마술봉): 사용할 수 있는 특정 무기입니다. 각각은 attack() 및 defend() 메소드의 자체 구현을 가지고 있습니다. 예를 들어 검은 근거리 공격을 할 수 있고, 활은 장거리 공격을 할 수 있습니다.<br><br>
> 따라서 전략 패턴을 사용하면 캐릭터 자체를 변경하지 않고도 다양한 무기(전략) 간을 전환하여 캐릭터의 동작(컨텍스트)을 동적으로 변경할 수 있습니다. 이를 통해 캐릭터가 게임의 다양한 도전에 더 유연하고 적응할 수 있게 됩니다.<br><br>
> 출처: chatgpt :)
