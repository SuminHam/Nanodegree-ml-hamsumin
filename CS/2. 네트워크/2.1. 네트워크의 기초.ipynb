### 네트워크의 기초

네트워크: 노드와 링크가 서로 연결되어 있으며 리소스를 공유하는 집합.<br>
노드: 서버, 라우터, 스위치 등 네트워크 장치를 의미하고 링크는 유선 또는 무선을 의미함.

#### 2.1.1. 처리량과 지연시간

처리량(throughput): 링크 내에서 성공적으로 전달된 데이터 양. 얼만큼의 트래픽을 처리했는지 나타낸다.

* 트래픽: 특정 시점에 링크 내에 흐르는 데이터의 양.<br>
  처리량이 많아졌다 = 처리되는 트래픽이 많아졌다.<br>
  트래픽이 많아졌다 = 흐르는 데이터 양이 많아졌다.<br>

지연시간: 요청이 처리되는 시간. 어떤 메시지가 두 장치 사이를 왕복하는 데 걸린 시간.


#### 네트워크 토폴로지와 병목 현상

네트워크 토폴로지: 노드와 링크가 어떻게 배치되어 있는지에 대한 방식이자 연결 형태.



* 트리 토폴로지

  트리 형태로 배치한 네트워크 구성.

   ![](https://velog.velcdn.com/images/slr-09/post/8c722f44-83b5-4afe-9546-765c65981657/image.png)

   - 노드의 추가, 삭제가 쉬우며 특정 노드에 트래픽이 집중될 때 하위 노드에 영향을 끼칠 수 있음.


* 버스 토폴로지

  중앙 통신 회선 하나에 여러 개의 노드가 연결되어 공유하는 네트워크 구성. LAN에서 유용하다.

    ![](https://images.edrawsoft.com/kr/articles/edrawmax/topology/bus-topology.png)

    - 설치 비용이 적고 신뢰성이 우수하며 중앙 통신 회선에 노드를 추가하거나 삭제하기 쉽다.
    - 스푸핑이 가능한 문제점이 있다.


* 스타 토폴로지
  
  중앙에 있는 노드에 모두 연결된 네트워크 구성.

  ![](https://blog.kakaocdn.net/dn/cpHmqy/btqJ8eNfZvH/yMejYUWOa0W7MnI6OgM9wk/img.png)

  - 노드를 추가하거나 에러를 탐지하기 쉽고 패킷의 충돌 발생 가능성이 적음. 어떠한 노드에 장애가 발생해도 쉽게 에러를 잡을 수 있으며 장애 노드가 중앙 노드가 아닐 경우 다른 노드에 영향을 끼치는 것이 적음.
  - 중앙 노드에 장애가 발생하면 전체 네트워크를 사용할 수 없고 설치 비용이 고가임.


* 링형 토폴로지
  
  각각의 노드가 양 옆의 두 노드와 연결하여 전체적으로 고리처럼 하나의 연속된 길을 통해 통신.

  ![](https://velog.velcdn.com/images/rlaghdtlr012/post/1585852e-b210-44e9-b603-42d9158a6412/image.png)

  - 노드 수가 증가되어도 네트워크상의 손실이 거의 없고 충돌이 발생되는 가능성이 적고 노드의 고장 발견을 쉽게 찾음.
  - 네트워크 구성 변경이 어렵고 히선에 장애가 발생하면 전체 네트워크에 영향을 크게 끼치는 단점이 있음.
  

* 매시 토폴로지
  
  그물망처럼 연결되어 있는 구조.

  ![](https://thebook.io/img/080326/073.jpg)

  - 한 단말 장치에 장애가 발생해도 여러 개의 경로가 존재하므로 네트워크를 계속 사용할 수 있고 트래픽도 분산 처리가 가능.
  - 노드의 추가가 어렵고 구축 비용과 운용 비용이 고가임.



* 병목 현상<br>
  전체 시스템의 성능이나 용량이 하나의 구성 요소로 인해 제한을 받는 현상.
  

### 2.1.3. 네트워크의 분류

- LAN: 근거리 통신망. 전송 속도가 빠르고 혼잡하지 않음.
- MAN: 대도시 지역 네트워크. 평균의 속도, LAN보다는 혼잡함.
- WAN: 광역 네트워크. 속도는 낮고 MAN보다 혼잡함.


### 2.1.4. 네트워크 성능 분석 명령어

- ping<br>
  트워크 상의 다른 시스템에게 ICMP 에코 요청을 보내고 응답 시간을 테스트 할 때 사용. 해당 노드의 패킷 수신 상태와 도달하기까지 시간 등을 알 수 있음.
  ping [ip 혹은 도메인]

- nslookup<br>
  DNS(Domain Name System) 레코드 조회를 할 때 사용하는 명령어. 도메인을 IP 주소로 변환할 때 사용한다. 
  nslookup google.com

- tracert<br>
  내 컴퓨터에서 해당 사이트의 서버까지 네트워크 경로를 추적.
  ICMP 에코 요청 메시지를 사용하고, 목적지까지 경로를 찾기 시작하는데, 이 때 각각의 에코 요청은 Time to Live 값을 가짐.
  이후 첫번째 TTL 값을 1로 지정하고 패킷이 라우터를 지날 때마다 값이 1씩 감소.<br>
  TTL 값이 0이 되면 해당 라우터는 패킷을 폐기하고 원본 주소로 ICMP “시간 초과” 메시지를 반환.<br> 
  이런식으로 네트워크 노드(라우터)를 통과할 때마다 발생하는 지연시간을 체크할 수 있고, 경로를 따라가는 과정을 내가 확인할 수 있다. 각 단계별 경로를 파악할 수 있기 때문에 네트워크 성능 문제를 진단하거나 특정 지점에서 발생할 수 있는 문제를 확인할 수 있다.
