## 2.2. TCP/IP 4계층 모델



#### 2.2.1. 계층 구조

* TCP/IP
  
  4계층. 애플리케이션 + 전송 + 인터넷 + 링크

  - 애플리케이션 계층
    
    FTP, HTTP, SSH, SMTP, DNS 등 응용 프로그램이 사용되는 프로토콜 계층. 서비스를 실질적으로 제공한다.

  - 전송 계층
    
    송신자와 수신자를 연결하는 통신 서비스를 제공.<br>
    연결 지향 데이터 스트림 지원, 흐름 제어를 제공할 수 있으며 애플리케이션과 인터넷 계층 사이의 데이터가 전달될 때 중계 역할을 함.<br>
    대표적으로 TCP와 UDP가 있다.<br>
    
    - TCP는 가상회선 패킷 교환 방식을 사용.
        ![](https://blog.kakaocdn.net/dn/cCDjcF/btqQjmZ5LjH/3WXxXEXoouAbiUJAMfJ5ok/img.png)<br>
        각 패킷에 가상회선 식별자가 포함되며 모든 패킷을 전송하면 가상회선이 해제되고 해킷들은 전송된 순서대로 도착.


    - UDP는 데이터크램 패킷 교환 방식을 사용.
        ![](https://woovictory.github.io/img/datagram_packet.png)<br>
        패킷이 독립적으로 이동하며 최적의 경로를 선택ㄷ하여 가는데, 하나의 메시지에서 분할된 여러 패킷은 서로 다른 경로로 전송될 수 있으며 도착한 순서가 다를 수 있음.

    * TCP 연결 성립 과정
      
      - 3-웨이 핸드셰이크

        클라이언트와 서버가 통신할 때 다음과 같은 세 단계의 과정을 거침.

        ![](https://blog.kakaocdn.net/dn/ZxCvx/btrtRXUCSQc/iEPeloKQbQ8KJPJu3BFKKK/img.png)<br>

        1) syn 단계<br>
          클라이언트는 서버에 클라이언트의 ISN을 담아 SYN을 보냄.<BR>
          ISN은 새로운 TCP 연결의 첫 번째 패킷에 할당된 임의의 시퀀스 번호를 말하며, 이는 장치마다 다름.

        2) SYN + ACK 단계
          서버는 클라이언트의 SYN을 수신하고 서버의 ISN을 보내며 승인번호로 클라이언트의 ISN+1을 보냄.

        3) ACK 단계
          클라이언트는 서버의 ISN+1한 값인 승인번호를 담아 ACK를 서버에 보냄.

      TCP는 이 과정이 있어 신뢰성이 있는 계층이라고 하며, UDP는 이 과정이 없어 신뢰성이 없는 계층이라고 함.


    * TCP 연결 해제 과정

      - 4-웨이 핸드셰이크

        TCP가 연결을 해제할 때는 이 과정이 발생.

        ![](https://raw.githubusercontent.com/WeareSoft/tech-interview/master/contents/images/4-way-handshaking.png)<BR>

        1) 먼저 클라이언트가 연결을 닫으려고 할 때 FIN으로 설정된 세그먼트를 보냄. 그리고 클라이언트는 FIN_WAIT_1 상태로 들어가고 서버의 응답을 기다림.
       
        2) 서버는 클라이언트로 ACK라는 승인 세그먼트를 보냄. 그리고 CLOSE_WAIT 상태에 들어감. 클라이언트가 세그먼트를 받으면 FIN_WAIT_2 상태에 들어감.
        
        3) 서버는 ACK를 보내고 일정 시간 이후에 크라이언트에 FIN이라는 세그먼트를 보냄.

        4) 클라이언트는 TIME_WAIT 상태가 되고 다시 서버로 ACK를 보내서 서버는 CLOSED 상태가 됨. 이후 클라이언트는 어느 정도의 시간을 대기한 후 연결이 닫히고 클라이언트와 서버의 모든 자원의 연결이 해제 됨.

        - TIME_WAIT을 사용하는 이유

          지연 패킷이 발생할 경우를 대비하기 위해.<BR>
          두 장치가 연결이 닫혔는지 확인하기 위해.<BR>


  - 인터넷 계층

    장치로부터 받은 네트워크 패킷을 IP 주소로 지정된 목적지로 전송하기 위해 사용되는 계층.<br>
    IP, ARP, ICMP 등이 있으며 패킷을 수신해야 할 상대의 주소를 지정하여 데이터를 전달.<br>
    상대방이 ㅔㅈ대로 받았는지에 대해 보장하지 않는다.

  - 링크 계층

    전선, 광섬유, 무선 등으로 실질적으로 데이터를 전달하며 장치 간에 신호를 주고받는 규칙을 정하는 계층. 네트워크 접근 계층이라고도 한다.<br>
    물리 계층과 데이터 링크 계층으로 나누기도 하는데 물리 계층은 무선 LAN과 유선 LAN을 통해 0과 1로 이루어진 데이터를 보내는 계층을 말한다. 데이터 링크 계층은 이더넷 프레임을 통해 에러 확인, 흐름 제어, 접근 제어를 담당하는 계층을 말한다.
